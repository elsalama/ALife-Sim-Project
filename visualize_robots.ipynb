#### Visualize robot morphologies (shapes)

**Before evolution:** Run cells 1–3 to see a random robot.
**After evolution:** Run cells 1 and 6 to see your top 3 evolved robots (requires robot_0.npy etc from GA run).

```python
from matplotlib import pyplot as plt
plt.rcParams['figure.dpi'] = 150

def plot_robot(robot):
    masses = robot["masses"]
    springs = robot["springs"]
    fig = plt.figure(figsize=(6, 6))
    for spring in springs:
        plt.plot([masses[spring[0], 0], masses[spring[1], 0]], [masses[spring[0], 1], masses[spring[1], 1]], color="blue", alpha=0.5)
    plt.scatter(masses[:, 0], masses[:, 1], color="red")
    plt.axis("equal")
    plt.show()
```

```python
from robot import sample_robot
```

```python
robot = sample_robot()
plot_robot(robot)
```
